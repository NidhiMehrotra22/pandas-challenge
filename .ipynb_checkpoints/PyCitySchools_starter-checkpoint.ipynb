```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

## Local Government Area Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average maths score 

* Calculate the average reading score

* Calculate the percentage of students with a passing maths score (50 or greater)

* Calculate the percentage of students with a passing reading score (50 or greater)

* Calculate the percentage of students who passed maths **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
school_total=school_data_complete["school_name"].nunique ()
school_total
```

```python
student_total=school_data_complete["Student ID"].nunique()
student_total
```

```python
budget_total= school_data_complete["budget"].sum()
budget_total
```

```python
math_aver= school_data_complete["maths_score"].mean()
math_aver
```

```python
read_aver=school_data_complete["reading_score"].mean()
read_aver
```

```python
math_pass= len(school_data_complete.loc[school_data_complete["maths_score"]>=50, :])
math_pass
```

```python
math_pass_percent= math_pass/student_total*100
math_pass_percent
```

```python
read_pass= len(school_data_complete.loc[school_data_complete["reading_score"]>=50, :])
read_pass
```

```python
read_pass_percent= read_pass/student_total*100
read_pass_percent
```

```python
overall_pass= len(school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)])
overall_pass                                         
```

```python
overall_pass_percent= overall_pass/student_total*100
overall_pass_percent
```

```python
area_summary = pd.DataFrame({
    "Total number of unique schools": [school_total], 
    "Total students":[student_total],
    "Total budget": [budget_total],
    "Average maths score": [math_aver],
    "Average reading score": [read_aver],
    "% passing maths":[math_pass_percent],
    "% passing reading": [read_pass_percent],
    "overall passing":[overall_pass_percent]
                                       
                                       }) 
area_summary
```

## School Summary

* Create an overview table that summarises key metrics about each school, including:

```python
school_group_name = school_data_complete.groupby(["school_name"])
school_group_name
```

```python
school_type=school_group_name["type"].first()
school_type
```

```python
total_student_count= school_group_name["Student ID"].count()
total_student_count
```

```python
school_total_budget=school_group_name["budget"].first()
school_total_budget
```

```python
per_capita_spend= school_total_budget/total_student_count
per_capita_spend
```

```python
average_math_score= school_group_name["maths_score"].mean()
average_math_score
```

```python
average_read_score= school_group_name["reading_score"].mean()
average_read_score
```

```python
math_pass_score_school= school_data_complete.loc[school_data_complete["maths_score"]>=50, :].groupby(["school_name"])["maths_score"].count()
math_pass_score_school
```

```python
math_pass_score_school_percent=math_pass_score_school/total_student_count*100
math_pass_score_school_percent
```

```python
read_pass_score_school= school_data_complete.loc[school_data_complete["reading_score"]>=50, :].groupby(["school_name"])["reading_score"].count()
read_pass_score_school
```

```python
read_pass_score_school_percent=read_pass_score_school/total_student_count*100
read_pass_score_school_percent
```

```python
both_pass_score=(school_data_complete.loc[(school_data_complete["maths_score"]>=50) & (school_data_complete["reading_score"]>=50)]).groupby(["school_name"])["reading_score"].count()
both_pass_score
```

```python
passing_rate=(math_pass_score_school_percent+read_pass_score_school_percent)/2
passing_rate
```

```python
per_school_summary= pd.DataFrame({"School Type":school_type,
                                  "Total student Count": total_student_count,
                                  "Total School Budget":school_total_budget,
                                  "Per Capita Spending": per_capita_spend,
                                  "Average Math Score": average_math_score,
                                  "Average Reading Score": average_read_score,
                                  "Math Pass Score":math_pass_score_school,
                                  "Reading Pass Score":read_pass_score_school,
                                  "% Math Pass":math_pass_score_school_percent,
                                  "% Reading Pass":read_pass_score_school_percent,
                                  "Pass score for both Math and Reading": both_pass_score,
                                  "% Overall Passing":passing_rate
                                 })
per_school_summary
```

## Maths Scores by Year

* Create a table that lists the average maths score for students of each year level (9, 10, 11, 12) at each school.

  * Create a pandas series for each year. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
year9_maths_score = school_data_complete.loc[school_data_complete["year"]== 9].groupby(["school_name"])["maths_score"].mean()
year10_maths_score=school_data_complete.loc[school_data_complete["year"]== 10].groupby(["school_name"])["maths_score"].mean()
year11_maths_score=school_data_complete.loc[school_data_complete["year"]== 11].groupby(["school_name"])["maths_score"].mean()
year12_maths_score=school_data_complete.loc[school_data_complete["year"]== 12].groupby(["school_name"])["maths_score"].mean()
maths_scores_by_year= pd.DataFrame({"Year_9": year9_maths_score,
                                      "Year_10": year10_maths_score,
                                       "Year_11": year11_maths_score,
                                      "Year_12": year12_maths_score})
maths_scores_by_year

```

## Reading Score by Year

* Perform the same operations as above for reading scores

```python
year9_read_score = school_data_complete.loc[school_data_complete["year"]== 9].groupby(["school_name"])["reading_score"].mean()
year10_read_score=school_data_complete.loc[school_data_complete["year"]== 10].groupby(["school_name"])["reading_score"].mean()
year11_read_score=school_data_complete.loc[school_data_complete["year"]== 11].groupby(["school_name"])["reading_score"].mean()
year12_read_score=school_data_complete.loc[school_data_complete["year"]== 12].groupby(["school_name"])["reading_score"].mean()
reading_scores_by_year= pd.DataFrame({"Year_9": year9_read_score,
                                      "Year_10": year10_read_score,
                                       "Year_11": year11_read_score,
                                      "Year_12": year12_read_score})
reading_scores_by_year
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
spending_bins = [0, 585, 630, 645, 680]
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
per_school_summary.dtypes
```

```python
per_school_summary["Spending Ranges(Per Student)"]= pd.cut(per_school_summary["Per Capita Spending"],spending_bins, labels= group_names)
group_spend= per_school_summary.groupby(["Spending Ranges(Per Student)"])
spending_summary= group_spend.mean()
spending_summary[["Average Math Score","Average Reading Score","% Math Pass", "% Reading Pass","% Overall Passing"]]
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
size_bins = [0, 1000, 2000, 5000]
group_name = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
per_school_summary["School Size"]= pd.cut(per_school_summary["Total student Count"],size_bins, labels= group_name)
group_size= per_school_summary.groupby(["School Size"])
size_summary= group_size.mean()
size_summary[["Average Math Score","Average Reading Score","% Math Pass", "% Reading Pass","% Overall Passing"]]
```

## Scores by School Type

* Perform the same operations as above, based on school type

```python
group_type= per_school_summary.groupby(["School Type"])
type_summary= group_type.mean()
type_summary[["Average Math Score","Average Reading Score","% Math Pass", "% Reading Pass","% Overall Passing"]]
```
